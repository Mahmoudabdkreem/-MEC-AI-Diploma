Write a program that prints following shape

```
    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
```



````python
n = 5
i = 0
while i < n:
    j = 0
    while j < n - i - 1:
        print(" ", end="")
        j += 1
    j = 0
    while j < 2 * i + 1:
        print("*", end="")
        j += 1
    print()
    i += 1

i = n - 2
while i >= 0:
    j = 0
    while j < n - i - 1:
        print(" ", end="")
        j += 1
    j = 0
    while j < 2 * i + 1:
        print("*", end="")
        j += 1
    print()
    i -= 1



````
